#### Task 2

```python
t = np.linspace(0, 10, 51)
f = np.cos(t)
```

```python
plt.plot(t, f, color='green')

label_dict = {'fontsize': '14', 'color': 'grey'}
title_dict = {'fontsize': '16', 'color': 'DarkBlue'}

plt.title('График f(t)', fontdict=title_dict)
plt.xlabel('Значения t', fontdict=label_dict)
plt.ylabel('Значения f', fontdict=label_dict)

plt.axis([0.5, 9.5, -2.5, 2.5])

plt.show()
```

#### Task 3

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)

for i in range(len(ax.flatten())):
    ax.flatten()[i].set_title(f'График y{i}')
    
ax1.set_xlim([-5, 5])

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

```python
import pandas as pd

plt.style.use('fivethirtyeight')
```

```python
# plt.style.library
# to read more https://matplotlib.org/gallery/style_sheets/style_sheets_reference.html
```

```python
data_df = pd.read_csv('creditcard.csv')
data_df.head()
```

```python
data_df.describe()
```

```python
data_df['Class'].value_counts()
```

```python
p = plt.bar(['Normal', 'Fraud'], data_df['Class'].value_counts(), color='lightseagreen')

plt.ylabel('Transactions')
plt.title('Grouped Transactions: Normal/Fraud')

plt.show()
```

```python
p = plt.bar(['Normal', 'Fraud'], data_df['Class'].value_counts(), log=True, color='lightseagreen')

plt.ylabel('Transactions')
plt.title('Grouped Transactions: Normal/Fraud. Log form.')

plt.show()
```

```python
plt.hist(data_df[data_df['Class'] == 1]['V1'], bins=20,  alpha=0.5, color='r',  density=True, label='Fraud')
plt.hist(data_df[data_df['Class'] == 0]['V1'], bins=20,  alpha=0.5, color='grey', density=True, label='Normal')

plt.xlabel('Class') # просили такое название, но вроде тут уместнее 'V1' по оси
plt.ylabel('Fraction')
plt.legend()

plt.show()
```
